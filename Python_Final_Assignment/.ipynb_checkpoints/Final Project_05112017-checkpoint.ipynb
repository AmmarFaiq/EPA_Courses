### 2.1. Carbon Dioxide Emissions

In this part, as it is stated above, the purpose is to find correlations between carbon dioxide emission and defined indicators that can be related to the emission levels. Chosen indicators are "Renewable Energy Use" and "Fossil Energy Use". Our pre-assumption is that the existed carbon dioxide emission levels are highly correlated with the type of energy use. Another point that needs to be explained is the main purpose. We try to examine these climate change indicators because we believe that at the end there may be some useful conclusions especially for the countries that seek for evidence in this realm. We aim that the conclusions deduced from this analysis are going to be useful for the countries that have not implemented any policy for climate change, and are currently in the process of policy development for global climate change goals. 

With this purposes and beliefs, the starting point has been chosen as OECD countries. For carbon dioxide emissions, each OECD country's performance is examined. The code provided below loads the data from excel files which are provided by World Bank. Then, filters the related countries, handles the "NA" values, shapes the data and returns it in a final table. 

The code below involves a function called "load_filter_OECD". This function requires one input as the path of excel files for WB data. While using the code, please do not forget to update the path for excel files. This function is proper to use only for World Bank data's excel file structure.

```python
def load_filter_OECD(filePath):

    #defining OECD members, and individual countries
    listOECD = ("OECD members","Australia","Austria","Belgium","Canada","Chile","Czech Republic","Denmark","Estonia","Finland","France","Germany","Greece","Hungary","Iceland","Ireland","Israel","Italy","Japan","Korea, Rep.","Latvia","Luxembourg","Mexico","Netherlands","New Zealand","Norway","Poland","Portugal","Slovak Republic","Slovenia","Spain","Sweden","Switzerland","Turkey","United Kingdom","United States") 
    
    #reading the excel file
    df = pd.read_excel(filePath, sheetname='Data', skiprows=3)

    #defining number of rows
    rowNum = df.shape[0]
    #defining number of columns
    columnNum = df.shape[1]

    #filtering OECD countries
    for i in range(rowNum-1,-1,-1):
        if not (df.iloc[i][0] in listOECD):
            df = df.drop(i, axis=0)
    
    #re-index the dataframe
    df = df.reset_index(drop=True)
    
    #handle the NA values
    df = df.dropna(axis=1)
    
    df = df.drop('Country Code', axis=1)
    df = df.drop('Indicator Code', axis=1)
    
    return df


#Please, do not forget to update the path for your computer.
#creating CO2 emission dataframes.
pathCO2 = 'co2EmissionData.xls'
CO2_df = load_filter_OECD(pathCO2)

#Please, do not forget to update the path for your computer.
#creating Renewable Energy Use dataframes
pathREN = 'renewableEnergyData.xls'
REN_df = load_filter_OECD(pathREN)

#Please, do not forget to update the path for your computer.
#creating Fossil Fuel Use dataframes
pathFEN = 'fossilEnergyData.xls'
FEN_df = load_filter_OECD(pathFEN)

#merge the three different data
frames = [CO2_df, REN_df, FEN_df]
result = pd.concat(frames, axis = 0)

#shape the data
melted = pd.melt(result, id_vars = result.columns.values[len(result.columns.values)-2:len(result.columns.values)], value_vars = result.columns.values[0:len(result.columns.values)-2])
finalData = melted.pivot_table(values='value', index=['variable'], columns = ['Indicator Name', 'Country Name'])

#equalize the number of rows by dropping NA's
finalData = finalData.dropna(axis=0)

#shows a part of finalized table.
finalData.head()
```

By using "finalData" table, one can easily reach statistics related to the data. The provided examples include average value of CO2 emissions by OECD members in general, maximum level of fossil fuel energy consumption percentage reached by Germany, and minimum level of renewable energy consumption reached by Netherlands. 

Also there are two functions provided in the code. First one is used to find the year of a maximum value of an indicator in a country. Second one is the opposite, it is used to find the minimum value.

```python
finalData.head()

print(np.mean(finalData['CO2 emissions (kt)']['OECD members']))
print(max(finalData['Fossil fuel energy consumption (% of total)']['Germany']))
print(min(finalData['Renewable energy consumption (% of total final energy consumption)']['Netherlands']))

#function to find the year of a maximum value
def findYearMax(key1, key2, data):
    #define the max value
    maxVal = max(data[key1][key2])
    #look for the max value to find the related year
    for i in range(0,len(data[key1][key2])):
        if data[key1][key2][i] == maxVal :
            return (finalData.iloc[i].name)

#function to find the year of a minimum value
def findYearMin(key1, key2, data):
    #define the max value
    maxVal = min(data[key1][key2])
    #look for the max value to find the related year
    for i in range(0,len(data[key1][key2])):
        if data[key1][key2][i] == maxVal :
            return (finalData.iloc[i].name)

#Examples for the use of findYearMax and findYearMin functions.
indicator1 = 'Fossil fuel energy consumption (% of total)'
country1 = 'Turkey'
maxFossilYear = findYearMax(indicator1, country1, finalData)
print('The maximum amount of ' + indicator1 + ' of ' + country1 +' was in ' + maxFossilYear)

indicator2 = 'Renewable energy consumption (% of total final energy consumption)'
country2 = 'United States'
minRenewableYear = findYearMin(indicator2, country2, finalData)
print('The minimum amount of ' + indicator2 + ' of ' + country2 +' was in ' + minRenewableYear)

indicator3 = 'CO2 emissions (kt)'
country3 = 'Switzerland'
minCO2 = findYearMin(indicator3, country3, finalData)
print('The minimum amount of ' + indicator3 + ' of ' + country3 +' was in ' + minCO2)
```

### 2.2. Methane Emissions

Two different database sources and type was used related to Methane gas emission analysis. First, We used World bank library for python, called 'wbdata' which take an updated data directly from The World Bank Database. Two data series was taken from the World Bank database, CH4 gas emission (kt) for predictor, and livestock production index as the first indicator. 

The wbdata library used its "Indicator Code" or "Indicator Name" as a parameter for getting the dataframe. The data_date should also be included as this analysis wanted to analyze from 1994 and beyond. Country code was also need to be included in the parameter. We used OECD members for the data form The World Bank which had an "OED" country code.

For the second database, we used OECD database about rice crop production and municipal waste in OECD countries for the second and third indicator. The data was downloaded to the working directory from the OECD website in CSV format and renamed to WASTEMUN.csv and CROPYIELD_RICE.csv to make it easier to call for further analysis. 

- Data pre-processing was conducted for methane gas emission analysis first by loading the data series into dataframe. This step was done by code below:

```python
#set up the indicator name (just build up the dict if you want more than one)
wbdata_ch4 = {'EN.ATM.METH.KT.CE':'CH4 emissions (kt)'}

wbdata_livestock = {'AG.PRD.LVSK.XD': 'Livestock production index (2004-2006 = 100)'}

#set the timeframe of data to be taken
data_date = (datetime.datetime(2016, 1, 1), datetime.datetime(1994, 1, 1))

#grab indicators above for countires above and load into data frame
df_wbdata_ch4 = wbdata.get_dataframe(wbdata_ch4, country='OED', convert_date=False, data_date = data_date)
df_wbdata_livestock = wbdata.get_dataframe(wbdata_livestock, country='OED', convert_date=False, data_date = data_date)

# change column name of WB data
df_wbdata_ch4['YEAR'] = df_wbdata_ch4.index.values
df_wbdata_ch4.sort_index(inplace=True)
df_wbdata_livestock['YEAR'] = df_wbdata_livestock.index.values
df_wbdata_livestock.sort_index(inplace=True)

#grab OECD data from csv file
csv_oecd_waste = pd.read_csv('WASTEMUN.csv')
csv_oecd_crop = pd.read_csv('CROPYIELD_RICE.csv')
```

- Since two database have different formatting, the OECD csv dataframes and wbdata dataframes pre-processed differently. Above code already changed several things regarding the column name of wbdata dataframes about livestock and CH4 emissions. The next code changed some formatting regarding OECD datafiles to make it consistent with each other. Two factors was considered important for both of the dataframes, which are YEAR and its VALUE (values column names differ depending of the indicator names). 

```python

def preprocessing_data_oecd_csv(df):
    
    # define the new column name for 'Value' in OECD database
    value = df.INDICATOR[1] + '(' + df.MEASURE[1] + ')'
    
    # pick data which only contains overall OECD members (not single state member)
    df = df[df['LOCATION'] == ('OECD')]
    
    # change column names of the dataframe
    df = df.rename(columns = {'TIME':'YEAR', 'Value':  value})
    
    # drop several unneeded columns
    df = df.drop(df.columns[[0,1,2,3,4,7]], axis=1)
    
    #change the order of the table
    df = df[[value,'YEAR']]
    
    return df

csv_oecd_crop2 = preprocessing_data_oecd_csv(csv_oecd_crop)
csv_oecd_waste2 = preprocessing_data_oecd_csv(csv_oecd_waste)
```

- Two of these dataframes also needs to be merged to make further analysis easier. This step was done in the next line of code, which also contained function to remove missing values. 

```python
def matchingdatayear(df1,df2):
    
    # equalize "type" of year for both of dataframes
    df1[['YEAR']] = df1[['YEAR']].astype(str).astype(int)
    df2[['YEAR']] = df2[['YEAR']].astype(str).astype(int)
    
    # merge two dataframe
    result = pd.merge(df1,df2, on='YEAR', how='inner')
    
    # drop some missing values
    result = result.dropna() 
    
    return result

# merge data from oecd database
mergedata= matchingdatayear(csv_oecd_crop2,csv_oecd_waste2)

# merge data from The World Bank Database
mergedata2= matchingdatayear(df_wbdata_ch4,df_wbdata_livestock)

# merge all of the data into one dataframe of dinal ch4 emission data
mergedata_ch4_final = matchingdatayear(mergedata,mergedata2)

# changing the order of the dataframe 
cols = list(mergedata_ch4_final.columns.values)

#sorting coloumn so that year will always be the first column on the right
mergedata_ch4_final =  mergedata_ch4_final[[cols[1],cols[3],cols[0],cols[2],cols[4]]]

#rename some of the coloumn names to make it more readable
mergedata_ch4_final = mergedata_ch4_final.rename(columns = {'CROPYIELD(TONNE_HA)':'Crop Production (Tonnes/Hectare)', 'WASTEMUN(KG_CAP)':  'Municipality Waste (Kilograms/capita)','YEAR':'Year'})
```

- The dataframe for CH4 or methane emissions already set and can be used for further analysis. There are few assumption regarding preprocessing data, these are:

    1. Since the missings data occur at the end and at the first of the series of data, interpolation is not an option for fill in the missing values, therefore we delete or drop the missing values was used to filter out those missing data

    2. The year 1994 to 2012 was also chosen for analysis simply because all of the indicators and predictors have complete datasets during these period.

```python
# Checking the head of the data we loaded.
WDI_data.head()
```

After we loaded the file, we have to filter the data according to our need. In this case, we need to focus on OECD members and get 3 indicators such as fossil fuel energy consumption, fertilizer consumption, and nitrous oxide emissions. But before that, we will define a filtering function that we will reuse in the next section of data preparation.

```python
def filter_data(dataframe, country, indicator1, indicator2, indicator3):
    WDI_data_filtered = WDI_data[(dataframe['Country Name'] == country) &
                            ((dataframe['Indicator Name'] == indicator1) |
                            (dataframe['Indicator Name'] == indicator2) |
                            (dataframe['Indicator Name'] == indicator3))]
    
    return WDI_data_filtered
```

```python
WDI_data_filtered = filter_data(WDI_data, 'OECD members', 'Fossil fuel energy consumption (% of total)', 'Fertilizer consumption (kilograms per hectare of arable land)', 'Nitrous oxide emissions (thousand metric tons of CO2 equivalent)')
```

```python
# Show WDI_data_filtered
WDI_data_filtered.head()
```

After we get filtered DataFrame, we will define the function below to melt, pivot, remove the NaN data. We will later reuse this function in the second section of Nitrous Oxide data preparation.

```python
def melt_pivot_cleanna(dataframe):
    # Melt the data where we identified the identification variable and value variables in the data.
    WDI_data_melt = pd.melt(dataframe, id_vars = dataframe.columns.values[0:3], value_vars = dataframe.columns.values[4:len(dataframe.columns.values)])
    
    # After we get the structure like above, we need to pivot the data where we set the year as index and 3 main contributors as the value
    WDI_data_final = WDI_data_melt.pivot_table(values='value', index=['variable', 'Country Name', 'Country Code'], columns = ['Indicator Name'])
    
    # Some values has NaN values. Use dropna to drop those values
    WDI_data_final = WDI_data_final.dropna(axis=0, how='any')
    
    return WDI_data_final
```

```python
WDI_data_final = melt_pivot_cleanna(WDI_data_filtered)
```

```python
WDI_data_final.head()
```

```python

# define x-axis and y-axis using dictionary that made from final dataframe of "mergedata_ch4_final" to be parameter for creating the plot
dict_scatter_ch4 = {'x':np.array(mergedata_ch4_final['Year']), 'y':np.array(mergedata_ch4_final['CH4 emissions (kt)'])}
dict_scatter_livestock = {'x':np.array(mergedata_ch4_final['Year']), 'y':np.array(mergedata_ch4_final['Livestock production index (2004-2006 = 100)'])}
dict_scatter_munwas = {'x':np.array(mergedata_ch4_final['Year']), 'y':np.array(mergedata_ch4_final['Municipality Waste (Kilograms/capita)'])}
dict_scatter_cropprod = {'x':np.array(mergedata_ch4_final['Year']), 'y':np.array(mergedata_ch4_final['Crop Production (Tonnes/Hectare)'])}

# declaration to make 4 subplot of a predictor and three indicators
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,12))

# define polyfit and polyval methane data with degree of three
coef = np.polyfit(dict_scatter_ch4['x'],dict_scatter_ch4['y'],3)
fit_line = np.polyval(coef,dict_scatter_ch4['x'])

# first subplot that contains scatter plot of methane gas and its regression line in time series (year)
ax[0,0].plot(dict_scatter_ch4['x'],dict_scatter_ch4['y'],'bo')
ax[0,0].plot(dict_scatter_ch4['x'], fit_line, label = 'Regression Line', color='red')
ax[0,0].set_title('CH4 Emissions')
ax[0,0].set_xlabel('Year')
ax[0,0].set_ylabel('kt')

# redefining x-axis value (time) as a rounded value since it may appear as a decimal value
xint = range(min(dict_scatter_ch4['x']), math.ceil(max(dict_scatter_ch4['x']))+1)
ax[0,0].set_xticks(xint[::2])

# define polyfit and polyval crop production data with degree of three
coef = np.polyfit(dict_scatter_cropprod['x'],dict_scatter_cropprod['y'],3)
fit_line = np.polyval(coef,dict_scatter_cropprod['x'])

# second subplot that contains scatter plot of crop production and its regression line in time series (year)
ax[0,1].plot(dict_scatter_cropprod['x'],dict_scatter_cropprod['y'],'bo')
ax[0,1].plot(dict_scatter_cropprod['x'], fit_line, label = 'Regression Line', color='red')
ax[0,1].set_title('Crop Production')
ax[0,1].set_xlabel('Year')
ax[0,1].set_ylabel('Tonnes / Hectare')

# redefining x-axis value (time) as a rounded value since it may appear as a decimal value
xint = range(min(dict_scatter_cropprod['x']), math.ceil(max(dict_scatter_cropprod['x']))+1)
ax[0,1].set_xticks(xint[::2])

# define polyfit and polyval livestock production data with degree of three
coef = np.polyfit(dict_scatter_livestock['x'],dict_scatter_livestock['y'],3)
fit_line = np.polyval(coef,dict_scatter_livestock['x'])

# third subplot that contains scatter plot of livestock production and its regression line in time series (year)
ax[1,0].plot(dict_scatter_livestock['x'],dict_scatter_livestock['y'],'bo')
ax[1,0].plot(dict_scatter_livestock['x'], fit_line, label = 'Regression Line', color='red')
ax[1,0].set_title('Livestock production')
ax[1,0].set_xlabel('Year')
ax[1,0].set_ylabel('Production index (2004-2006 = 100)')

# redefining x-axis value (time) as a rounded value since it may appear as a decimal value
xint = range(min(dict_scatter_livestock['x']), math.ceil(max(dict_scatter_livestock['x']))+1)
ax[1,0].set_xticks(xint[::2])

# define polyfit and polyval municipality waste data with degree of three
coef = np.polyfit(dict_scatter_munwas['x'],dict_scatter_munwas['y'],3)
fit_line = np.polyval(coef,dict_scatter_munwas['x'])

# forth subplot that contains scatter plot of municipality waste and its regression line in time series (year)
ax[1,1].plot(dict_scatter_munwas['x'],dict_scatter_munwas['y'],'bo')
ax[1,1].plot(dict_scatter_munwas['x'], fit_line, label = 'Regression Line', color='red')
ax[1,1].set_title('Municipality Waste')
ax[1,0].set_xlabel('Year')
ax[1,0].set_ylabel('Kilograms / capita')

# redefining x-axis value (time) as a rounded value since it may appear as a decimal value
xint = range(min(dict_scatter_munwas['x']), math.ceil(max(dict_scatter_munwas['x']))+1)
ax[1,1].set_xticks(xint[::2])
```

### 3.3.4. Training and Testing the Data for Machine Learning

The first step to do this is by doing data classification for predicted data and predictor. In this case, the factor we want to predict is Nitrous Oxide emissions while the factor for prediction is fossil fuel emissions and fertilizer to be consistent with data from NASA.

```python
y = WDI_data_final['Nitrous oxide emissions (thousand metric tons of CO2 equivalent)']
```

```python
# Identify the list of variable inside the arguments
l = ['Fossil fuel energy consumption (% of total)', 'Fertilizer consumption (kilograms per hectare of arable land)']
X = WDI_data_final[l]
```

```python
y2 = WDI_data_final['Nitrous oxide emissions (thousand metric tons of CO2 equivalent)']
```

```python
X2 = WDI_data_final[['Fossil fuel energy consumption (% of total)']]
```

In below section, we will separate the training and testing data. As identified in data preparation before separating the training and testing data. We have to do library initialization first of scikit learn by running 'from sklearn.model_selection import train_test_split'. For the test_size variable, it is to define how many test variable we would like to take from the data to make a prediction while the train size is simply by substracting the test size from 1. Train size is the variable to teach the model from available data and the program will do this automatically after we define the test size as below code.

```python
# Define train and test variable for linear regression model 1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=101)
```

```python
# Define train and test variable for linear regression model 2
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.7, random_state=101)
```

### 3.3.5. Training the Model Using Machine Learning Method

We will use Scikit Learn library to generate linear model. The initialization of Scikit learn library has been done in the data preparation section above. In this section, we will use teach the Scikit using X_train and y_train set that we have defined previously. Then, we will generate the linear regression model and show the coefficient for interception and slope. As defined in the comment on the particular cell, there are 2 linear model which are the linear regression model between nitrous oxide emissions and fossil fuel combustion and between nitrous oxide emissions and fertilizer consumption. lm will generate a model between nitrous oxide emissions and combination of 2 factors such as fertilizer consumption and fossil fuel consumption while lm2 generates model between nitrous oxide emissions and fossil fuel consumption which has the highest correlation to nitrous oxide emissions.

```python
# Initialize linear regression model called lm
# lm is linear regression model where we include 2 variables such as fossil fuel combustions and fertilizer consumption
# while lm2 is linear regression model where we include only the factor with highest correlation

# lm is a model between nitrous oxide emissions and combinations of 2 factors which are fossil fuel consumption and fertilizer consumption
lm = LinearRegression()

# lm2 is a model between nitrous oxide emissions and fossil fuel consumption only.
lm2 = LinearRegression()
```

```python
# Train/fit first linear regressions model on the training data
lm.fit(X_train, y_train)
```

```python
# Train/fit linear regressions model 2 on the training data
lm2.fit(X2_train, y2_train)
```

```python
# Checking the coefficient of regression model lm
print('Coefficients: \n', lm.coef_)
```

```python
# Checking the coefficient of regression model lm2
print('Coefficients: \n', lm2.coef_)
```

### 3.3.7. Model Evaluation

After we successfully generated the plot in previous section, we will show some statistical metric for evaluations purpose which are mean absolute error, mean squared error, and root of mean squared error which indicated the average of errors or deviations in the model. In other words, we show the difference between y test and predictions. Sometimes, the number will show the result that's a bit hard to understand because it depends on the scale of particular variables. Hence, in this section, we will show distribution plot of our result to have quick understanding of quality of our model. Model with good fit will produce normal distribution plotting result as shown below. As a point of attention, because our data only spans for around 10 years, it will produce few histogram as shown in normal distribution plot below. In the end, we conclude our model has good fit although it has few data points only.

```python
from sklearn import metrics

# Show the mean squared error according to explanation above for statistical purpose
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Plotting the residuals of the distributions
sns.distplot((y_test-predictions),bins=8);
```

Based on the plot above, we can see normal distribution in the residuals (y_test - predictions) so we can conclude we have a good model with a good fit.
