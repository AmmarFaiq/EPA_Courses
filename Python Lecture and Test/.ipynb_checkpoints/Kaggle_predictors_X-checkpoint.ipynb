```python
import pandas as pd
import numpy as np
df = pd.read_csv("final_X.csv")
```

```python
# from 15-01 to 16-04
date = df["Date"].unique()[:-1]
# from 15889 to ***
cus = df["Cus_cod"].unique()

cus_arr = np.repeat(cus, len(date))
date_arr = np.tile(date, len(cus))
id_date_cuscd = date_arr + np.repeat("-", len(date_arr)) + cus_arr.astype(str)
```

```python
df = df.drop("Cus_cod", axis = 1)
```

```python
X = pd.DataFrame(index = id_date_cuscd, columns = df.columns)
X.loc[:] = df
```
