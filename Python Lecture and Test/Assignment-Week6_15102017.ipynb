The file ```titanic.csv``` provides passenger information on the ship RMS Titanic that sank on her maiden voyage on April 15, 1912. 

**1.** Read the passenger information of the Titanic. Show an overview of _unique_ last names of the passenger list (so ignore the first names and the "Mr", "Mrs.", etc) **and** determine the number of different last names.

````python
# your code here
titanic = pd.read_csv('titanic.csv')
print (len(titanic['Name'].apply(lambda words: words.split(',')[0]).unique()))
len(titanic['Name'].apply(lambda words: words.split(',')[0]).unique())
````

**2.** Assuming that passengers with the same last name form a family, determine how big the largest family aboard the Titanic was.

````python
titanic['Last Name'] = titanic['Name'].apply(lambda words: words.split(',')[0])
titanic['Last Name'].value_counts().max()
````

**3.** Drop all the rows for which the Fare is 0 or NaN. Store the result.

````python
titanic_new = titanic[titanic['Fare'] != 0].dropna(axis=0, subset = ['Fare'])
````

````python
len(titanic_new)
````

**4.** Show the average Fare price per Passenger Class (Pclass) in a bar chart.

````python
# Group by Pclass
by_Pclass = titanic_new.groupby('Pclass')

# Plotting
by_Pclass.mean()['Fare'].plot(kind='bar')
plt.ylabel('Average Fare Price')
plt.title('Average Fare Price per P Class')
plt.show()
````

**5.** Plot a histogram of the passenger's age, grouped by sex. So the end-result should show separate  histograms for males and females.

````python
# Plotting
# Plotting Male's Age histogram 
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
titanic_new[titanic_new['Sex'] == 'male']['Age'].plot.hist()
plt.title("Male Passenger's Age")
plt.xlabel('Age')

# Plotting Female's Age histogram
plt.subplot(2,1,2)
titanic_new[titanic_new['Sex'] == 'female']['Age'].plot.hist(color='blue')
plt.title("Female Passenger's Age")
plt.xlabel('Age')

# To prevent overlapping plot
plt.tight_layout()
````
